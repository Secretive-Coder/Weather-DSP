```python
import pandas as pd
data = pd.read_csv(r"C:\Users\jangh\Downloads\weather.csv")
#Find all the unique "Wind Speed" values in the data.
print(data["Wind Speed_km/h"].unique())
```

```python
#Find the number of times when the weather is exactly clear.
clear_count = (data["Weather"]=="clear").sum()
print("Number of times 'clear' appear:", clear_count)
```

```python
#Find the number of times when the wind speed was exactly 4km/h.
wind_speed_4km = (data['Wind Speed_km/h']==4).sum()
print("Number of times 'Wind Speed' was exactly 4km/h:",wind_speed_4km)
```

```python
#Find out all the null values in the data.
print(data.isnull().sum())
```

```python
#Rename the column name 'Weather' of the dataframe to 'Weather Condition'.
data.rename(columns={"Weather":"Weather Condition"})
```

```python
#Mean of 'Visibility'
data_mean_visibility=data["Visibility_km"].mean()
print(data_mean_visibility)
```

```python
#Standard Deviation of 'Pressure'.
std_pressure=data["Press_kPa"].std()
print(std_pressure)
```

```python
#Variance of 'Relative Humidity'.
varience_hum=data["Rel Hum_%"].var()
print(varience_hum)
```

```python
#Find all instances when 'Snow' was recorded.
data_snow_count=data[data["Weather"]=="Snow"]
print(data_snow_count)
```

```python
#Find all the instances when 'Wind Speed is above 24' and 'Visibility is 25'.
wind_vis=data[(data["Wind Speed_km/h"]>24) & (data["Visibility_km"]==25)]
print(wind_vis)
```

```python
#Mean value of each column against each 'Weather Condition'.
mean_value = data.groupby('Weather').mean()
print(mean_value)
```

```python
#Minimum and Maximum value of each column against each 'Weather Condition'.
max_min_each=data.groupby("Weather").agg(['min','max'])
print(max_min_each)
```

```python
#Records where 'Weather Condition' is 'Fog'.
data_fog_count=data[data["Weather"]=="Fog"]
print(data_fog_count)
```

```python
#Find all instances when 'Weather is clear' or 'Visibility is above 40'.
wind_vis2=data[(data["Weather"]=="clear")|(data["Visibility_km"]>40)]
print(wind_vis2)
```

```python
#Find all instances when 'Weather is clear' and 'Relative Humidity' is greater than 50.
wind_vis3=data[(data["Weather"]=="clear")|(data["Rel Hum_%"]>50)]
print(wind_vis3)
```

```python
#Find all instances when 'Visibility is above 40'.
wind_vis2=data[(data["Visibility_km"]>40)]
print(wind_vis2)
```
